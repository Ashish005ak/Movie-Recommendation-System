```python
import ast
def cont(Text):
  l=[]
  for i in ast.literal_eval(Text):
    l.append(i['name'])
  return l
m['genres']=m['genres'].apply(cont)
```

```python
m.iloc[0]['genres']
```

```python
m.head()
```

```python
m['keywords'] = m['keywords'].apply(cont)
```

```python
m.head()
```

```python
def conv(Text):
  l=[]
  counter=0
  for i in ast.literal_eval(Text):
    if counter>3:
      l.append(i['name'])
    counter+=1
  return l
m['cast']=m['cast'].apply(conv)
```

```python
m.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
m['crew']=m['crew'].apply(fetch_director)
```

```python
m['overview']=m['overview'].apply(lambda x:x.split())
```

```python
def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
m['overview']=m['overview'].apply(remove_space)
```

```python
m['overview']
```

```python
m['cast'] = m['cast'].apply(remove_space)
m['crew'] = m['crew'].apply(remove_space)
m['genres'] = m['genres'].apply(remove_space)
```

```python
m.head()
```

```python
m['tags'] = m['overview'] + m['genres'] + m['keywords'] + m['cast'] + m['crew']
```

```python
m1=m[['movie_id','title','tags']]
```

```python
m1.head()
```

```python
m1['tags']=m1['tags'].apply(lambda x : " ".join(x))
```

```python
type(m1.iloc[0]['tags'])
```

```python
m1['tags']=m1['tags'].apply(lambda x : x.lower())
```

```python
m1.head()
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(Text):
  T=[]
  for i in Text.split():
    T.append(ps.stem(i))
  return " ".join(T)
m1['tags'] = m1['tags'].apply(stem)
```

```python
m1.iloc[0]['tags']
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
n = set(stopwords.words('english'))
def remove(Text):
  words = Text.split()
  ft = [i for i in words if i not in n]
  return " ".join(ft)
m1['tags']=m1['tags'].apply(remove)
```

```python
m1.iloc[0]['tags']
```

```python
m1.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
```

```python
vector = cv.fit_transform(m1['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
vector[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
m1[m1['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = m1[m1['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(m1.iloc[i[0]].title)
```

```python
recommend('Mulan')
```
